## Vos réponses?

1- Les résultats obtenus sont similaires et ce sont les bons résultats attendus. 
Les valeurs de λk et μk sont aussi importants vu qu'ils sont bien liés au nombre d'itérations et notamment au critère de convergence global. Et d'après l'expression du Lagrangien, on peut voir que si on augmente la valeur de μk, on accélère l'algorithme (multiplication à un terme au carré), contrairement à λk qui n'a pas beaucoup d'influence sur l'algorithme malgré son importance. Dans ce sens, μk permet de résoudre des problèmes complexes qui prennent beaucoup de temps si on applique d'autres méthodes de résolution de problèmes. 

2- On peut voir d'après l'algorithme que τ est lié à μk sous une condition sur la contrainte, et donc plus on augmente τ, plus on augmente aussi μk, et comme expliqué dans la question précédente, ceci permet de trouver une meilleure convergence à l'algorithme. Cependant, on ne peut pas choisir une valeur de τ très petite (car sinon on perd en précision), et on ne peut pas choisir une valeur de τ qui soit très grande. On doit donc bien choisir une valeur adaptée à notre problème.

3- Afin de résoudre des problèmes avec contraintes à la fois d'égalité et d'inégalité, on peut exprimer les inégalités sous formes d'égalités (en ajoutant de nouveaux termes par exemple), et comme ça on se ramène à un algorithme similaire à le nôtre mais avec plus de contraintes et plus de termes.
